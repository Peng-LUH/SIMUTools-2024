```python
from sapn.regions import regions
from sapn.objects.sa_transition_system import SATransitionSystem
from pprint import pprint

test_state_transitions_3 = [('s_0', 'r', 's_1'), 
                            ('s_1', 's', 's_2'), ('s_1', 'sb', 's_3'), ('s_1', 'em', 's_4'), 
                            ('s_2', 'sb', 's_5'), 
                            ('s_3', 's', 's_5'), ('s_3', 'p', 's_6'), ('s_3', 'em', 's_7'), 
                            ('s_4', 'sb', 's_7'), 
                            ('s_5', 'p', 's_8'),
                            ('s_6', 's', 's_8'), ('s_6', 'em', 's_9'),
                            ('s_7', 'p', 's_9'),
                            ('s_8', 'ac', 's_10'),
                            ('s_9', 'ac', 's_11'),
                            ('s_10', 'ap', 's_12'), ('s_10', 'rj', 's_13'),
                            ('s_11', 'ap', 's_14'), ('s_11', 'rj', 's_15'),
                            ('s_12', 'c', 's_16'),
                            ('s_13', 'rs', 's_17'),
                            ('s_14', 'c', 's_18'),
                            ('s_15', 'rs', 's_19'),
                            ('s_17', 'c', 's_20'),
                            ('s_19', 'c', 's_21')]

ts_3 = SATransitionSystem()
ts_3.add_transitions_batch(state_transitions=test_state_transitions_3)
ts_3.ts_view()
```

```python
import time

start_time = time.time()

discovered_minimal_regions, explored_multisets, iterations = regions.generate_all_minimal_regions_o(k=1, transition_system=ts_3)

end_time = time.time()
computation_time = end_time - start_time


print(f"Minimal Regions: ")
for m in discovered_minimal_regions:
    print(list(m.values()))
```

```python
print(f"\nNum of discovered minimal regions: {len(discovered_minimal_regions)}")
print(f"\nNum of explored multisets: {len(explored_multisets)}")
print(f"\nNum of Iterations: {iterations}")
print(f"\nComputation time: {computation_time}")
```
